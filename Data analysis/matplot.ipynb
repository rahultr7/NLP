```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [20,6,7,12,19]

##create a line plot

plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Label')
plt.title('Test plot')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [20,6,7,12,19]

plt.plot(x,y,color='red',linestyle=':',marker='o',linewidth=3,markersize=5)
plt.grid(True)
```

```python
##Multiple plot 
##Sample data

x = [1,2,3,4,5]
y1 = [20,6,7,12,19]
y2 = [3,7,2,10,8]

plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.plot(x,y2)

plt.subplot(1,2,2)
plt.plot(y1,y2)
```

```python
cat = ['A','B','C','D','E']
val = [3,4,5,2,1]

plt.bar(cat,val,color = "red")
plt.xlabel('category')
plt.ylabel('values')
plt.title('Bar chart')
```

```python
cat = ['A','B','C','D','E']
val = [3,4,5,2,1]
colors =['red','blue','green','yellow','purple']
explode = (0.2,0,0,0,0)

plt.pie(val,explode=explode,labels=cat,colors=colors,autopct='%1.1f%%',shadow=True)
```

```python
import pandas as pd

sales_data = pd.read_csv('sales_data.csv')
sales_data.head(5)
```

```python
sales_data.info()
```

```python
total_sales = sales_data.groupby('Product Category')['Total Revenue'].sum()
print(total_sales)
```

```python
total_sales.plot(kind="bar",color='red')
```

```python
sales_trend = sales_data.groupby('Date')['Total Revenue'].sum().reset_index()
plt.plot(sales_trend['Date'],sales_trend['Total Revenue'])
```
