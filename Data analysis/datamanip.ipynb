```python
data1 = data.fillna(1)
```

```python
data['Sales_fill']=data['Sales'].fillna(data['Sales'].mean())
data.head(5)
```

```python
data.dtypes
```

```python
##Renaming column

rename_data = data.rename(columns={'Date':'Sales date'})
rename_data.head(5)
```

```python
##Change Datatype

data['Value_new'] = data['Value'].fillna(data['Value'].mean()).astype(int)
data.head(3)
```

```python
data['New value'] = data['Value'].apply(lambda x:x*2)
data.head(3)
```

```python
## Grouping

grouped_data = data.groupby(['Product','Region'])['Value'].sum()
grouped_data
```

```python
aggr_data = data.groupby('Region')['Value'].agg(['sum','mean','count'])
aggr_data
```

```python
data1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value1': [1, 2, 3]})
data2 = pd.DataFrame({'Key': ['A', 'B', 'D'], 'Value2': [4, 5, 6]})
```

```python
data1
```

```python
#Merge data

pd.merge(data1,data2,on='Key',how='inner')
```

```python
#Merge data

pd.merge(data1,data2,on='Key',how='outer')
```

```python
#Merge data

pd.merge(data1,data2,on='Key',how='left')
```

```python
#Merge data

pd.merge(data1,data2,on='Key',how='right')
```
