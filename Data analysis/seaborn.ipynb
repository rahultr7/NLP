```python
import seaborn as sbn
```

```python
##Basic Plotting

tips = sbn.load_dataset('tips')
```

```python
##Scatter plot

import matplotlib.pyplot as mtb

sbn.scatterplot(x='total_bill',y='tip',data=tips)
mtb.title('Demo Seaborn')
```

```python
##Scatter plot

import matplotlib.pyplot as mtb

sbn.lineplot(x='size',y='total_bill',data=tips)
mtb.title('Demo Seaborn')
```
